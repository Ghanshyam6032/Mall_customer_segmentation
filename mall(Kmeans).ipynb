```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
df.head()
```

```python
kmeans=KMeans(n_clusters=5,random_state=42)
label=kmeans.fit_predict(df)
center=kmeans.cluster_centers_
```

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],c=label)
plt.scatter(center[:,0],center[:,1],marker='X')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('KMeans Cluster')
plt.show()
```

```python
scale=StandardScaler()
df_scale=scale.fit_transform(df)
```

```python
db=DBSCAN(eps=0.4,min_samples=5)
leb=db.fit_predict(df_scale)
```

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],c=leb,cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBscan Cluster')
plt.show()
```
